```python
import pandas as pd
import numpy as np
```

```python
series = pd.date_range(start = '2022-02-28' , end = '2022-03-28', freq='20S')
```

```python
df = pd.DataFrame(index = series)
df['count'] = np.random.choice([5,4,3,2,1,0], p = [0.01, 0.02, 0.03, 0.04, 0.5,0.74], size = (len(series)))
```

```python
def cumsum_day(frame):
    cumul_list = []
    day = frame.index[0].day
    for i in range(len(frame)):
        if day == frame.index[i].day:
            if len(cumul_list) == 0:
                cumul_list.append(frame['count'][i])
            else:
                cumul_list.append((cumul_list[i-1]) + (frame['count'][i]))
        else:
            day = frame.index[i].day
            cumul_list.append(frame['count'][i])
    frame['cumulative count (day)'] = cumul_list
    return frame
df = cumsum_day(df)
```
